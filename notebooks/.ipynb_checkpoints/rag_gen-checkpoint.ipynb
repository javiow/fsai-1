```python
# ê°ê´€ì‹ ì—¬ë¶€ íŒë‹¨ í•¨ìˆ˜
def is_multiple_choice(question_text):
    """
    ê°ê´€ì‹ ì—¬ë¶€ë¥¼ íŒë‹¨: 2ê°œ ì´ìƒì˜ ìˆ«ì ì„ íƒì§€ê°€ ì¤„ ë‹¨ìœ„ë¡œ ì¡´ì¬í•  ê²½ìš° ê°ê´€ì‹ìœ¼ë¡œ ê°„ì£¼
    """
    lines = question_text.strip().split("\n")
    option_count = sum(bool(re.match(r"^\s*[1-9][0-9]?\s", line)) for line in lines)
    return option_count >= 2


# ì§ˆë¬¸ê³¼ ì„ íƒì§€ ë¶„ë¦¬ í•¨ìˆ˜
def extract_question_and_choices(full_text):
    """
    ì „ì²´ ì§ˆë¬¸ ë¬¸ìì—´ì—ì„œ ì§ˆë¬¸ ë³¸ë¬¸ê³¼ ì„ íƒì§€ ë¦¬ìŠ¤íŠ¸ë¥¼ ë¶„ë¦¬
    """
    lines = full_text.strip().split("\n")
    q_lines = []
    options = []

    for line in lines:
        if re.match(r"^\s*[1-9][0-9]?\s", line):
            options.append(line.strip())
        else:
            q_lines.append(line.strip())

    question = " ".join(q_lines)
    return question, options

# í”„ë¡¬í”„íŠ¸ ìƒì„±ê¸°
def make_prompt_auto(text):
    if is_multiple_choice(text):
        question, options = extract_question_and_choices(text)
        prompt = (
                "ë‹¹ì‹ ì€ ê¸ˆìœµë³´ì•ˆ ì „ë¬¸ê°€ì…ë‹ˆë‹¤.\n"
                "ì•„ë˜ ì§ˆë¬¸ì— ëŒ€í•´ ì ì ˆí•œ **ì •ë‹µ ì„ íƒì§€ ë²ˆí˜¸ë§Œ ì¶œë ¥**í•˜ì„¸ìš”.\n\n"
                f"ì§ˆë¬¸: {question}\n"
                "ì„ íƒì§€:\n"
                f"{chr(10).join(options)}\n\n"
                "ë‹µë³€:"
                )
    else:
        prompt = (
                "ë‹¹ì‹ ì€ ê¸ˆìœµë³´ì•ˆ ì „ë¬¸ê°€ì…ë‹ˆë‹¤.\n"
                # "ì•„ë˜ ì£¼ê´€ì‹ ì§ˆë¬¸ì— ëŒ€í•´ ì •í™•í•˜ê³  ê°„ëµí•œ ì„¤ëª…ì„ ì‘ì„±í•˜ì„¸ìš”.\n\n"
                "ì•„ë˜ ì§ˆë¬¸ì— ëŒ€í•´ ì •ë‹µì˜ í•µì‹¬ í‚¤ì›Œë“œì™€ ì˜ë¯¸ë¥¼ ëª¨ë‘ í¬í•¨í•˜ì—¬ 3ë¬¸ì¥ ì´ë‚´ë¡œ ê°„ê²°í•˜ê²Œ ë‹µë³€í•˜ì„¸ìš”. êµ°ë”ë”ê¸° ì—†ì´ ìš”ì ë§Œ ëª…í™•í•˜ê²Œ ì‘ì„±í•˜ì„¸ìš”.\n\n"
                f"ì§ˆë¬¸: {text}\n\n"
                "ë‹µë³€:"
                )
    return prompt

# í›„ì²˜ë¦¬ í•¨ìˆ˜
def extract_answer_only(generated_text: str, original_question: str) -> str:
    """
    - "ë‹µë³€:" ì´í›„ í…ìŠ¤íŠ¸ë§Œ ì¶”ì¶œ
    - ê°ê´€ì‹ ë¬¸ì œë©´: ì •ë‹µ ìˆ«ìë§Œ ì¶”ì¶œ (ì‹¤íŒ¨ ì‹œ ì „ì²´ í…ìŠ¤íŠ¸ ë˜ëŠ” ê¸°ë³¸ê°’ ë°˜í™˜)
    - ì£¼ê´€ì‹ ë¬¸ì œë©´: ì „ì²´ í…ìŠ¤íŠ¸ ê·¸ëŒ€ë¡œ ë°˜í™˜
    - ê³µë°± ë˜ëŠ” ë¹ˆ ì‘ë‹µ ë°©ì§€: ìµœì†Œ "ë¯¸ì‘ë‹µ" ë°˜í™˜
    """
    # "ë‹µë³€:" ê¸°ì¤€ìœ¼ë¡œ í…ìŠ¤íŠ¸ ë¶„ë¦¬
    if "ë‹µë³€:" in generated_text:
        text = generated_text.split("ë‹µë³€:")[-1].strip()
    else:
        text = generated_text.strip()

    # ê³µë°± ë˜ëŠ” ë¹ˆ ë¬¸ìì—´ì¼ ê²½ìš° ê¸°ë³¸ê°’ ì§€ì •
    if not text:
        return "ë¯¸ì‘ë‹µ"

    # ê°ê´€ì‹ ì—¬ë¶€ íŒë‹¨
    is_mc = is_multiple_choice(original_question)

    if is_mc:
        # ìˆ«ìë§Œ ì¶”ì¶œ
        match = re.match(r"\D*([1-9][0-9]?)", text)
        if match:
            return match.group(1)
        else:
            # ìˆ«ì ì¶”ì¶œ ì‹¤íŒ¨ ì‹œ "0" ë°˜í™˜
            return "0"
    else:
        return text
```

```python
import torch
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
from langchain.schema import Document
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from langchain_community.vectorstores import Chroma
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnableLambda, RunnablePassthrough
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.document_loaders import TextLoader

documents = TextLoader("../data/cleaned_text_1.txt").load()
# ë¬¸ì„œë¥¼ ì²­í¬ë¡œ ë¶„í• 
def split_docs(documents, chunk_size=1000, chunk_overlap=20):
  text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
  docs = text_splitter.split_documents(documents)
  return docs

# docs ë³€ìˆ˜ì— ë¶„í•  ë¬¸ì„œë¥¼ ì €ì¥
docs = split_docs(documents)

# 1) HF ëª¨ë¸ê³¼ í† í¬ë‚˜ì´ì € ë¡œë“œ
model_id = "nlpai-lab/KULLM3"   # ì˜ˆì‹œ: ì›í•˜ëŠ” ëª¨ë¸ë¡œ êµì²´
tokenizer = AutoTokenizer.from_pretrained(model_id)
llm_model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    load_in_4bit=True,
    torch_dtype=torch.float16
)

MODEL_NAME = "jhgan/ko-sroberta-multitask"  # ë˜ëŠ” gte-multilingual-base, MiniLM ë“±

embeddings = HuggingFaceEmbeddings(
    model_name=MODEL_NAME,
    model_kwargs={
        "device": "cuda",  # í•„ìš” ì‹œ
    },
    encode_kwargs={
        "normalize_embeddings": True,
        "convert_to_tensor": True,   # â˜… NumPy ëŒ€ì‹  torch.Tensorë¡œ ë°›ê¸°
        "convert_to_numpy": False,   # â˜… ëª…ì‹œì ìœ¼ë¡œ ë„ê¸°
    },
)

db = Chroma.from_documents(
    documents=docs,
    embedding=embeddings,
    persist_directory="../db",   # Windowsë©´ ì ˆëŒ€ê²½ë¡œ ê¶Œì¥
    collection_name="law_docs",
)
```

```python
if tokenizer.pad_token_id is None:
    tokenizer.pad_token_id = tokenizer.eos_token_id
llm_model.generation_config.pad_token_id = tokenizer.pad_token_id
llm_model.generation_config.eos_token_id = tokenizer.eos_token_id

llm_model.eval()
torch.set_grad_enabled(False)
if torch.cuda.is_available():
    torch.backends.cuda.matmul.allow_tf32 = True
    torch.set_float32_matmul_precision("high")  # ì•ˆì „í•œ ì†ë„ í–¥ìƒ

# 2) í…ìŠ¤íŠ¸ ìƒì„± íŒŒì´í”„ë¼ì¸
gen_pipe = pipeline(
    task="text-generation",
    model=llm_model,
    tokenizer=tokenizer,
    device_map="auto",
    max_new_tokens=256,
    do_sample=False,
    temperature=0.2,
)

# 3) HF íŒŒì´í”„ë¼ì¸ì„ LangChain Runnableë¡œ ë˜í•‘
# def hf_generate(prompt: str) -> str:
#     out = gen_pipe(prompt, return_full_text=False)
#     return out[0]["generated_text"]

def hf_generate_direct(text: str) -> str:
    # ë¬¸ìì—´ ë³´ì¥(í˜¹ì‹œ ëª¨ë¥¼ PromptValue ë°©ì§€)
    s = text if isinstance(text, str) else getattr(text, "to_string", lambda: str(text))()
    inputs = tokenizer(s, return_tensors="pt")
    # ëª¨ë¸ ë””ë°”ì´ìŠ¤ë¡œ ì´ë™
    inputs = {k: v.to(llm_model.device) for k, v in inputs.items()}
    with torch.no_grad():
        outputs = llm_model.generate(
            **inputs,
            max_new_tokens=256,
            do_sample=False,
            temperature=0.2,
            eos_token_id=tokenizer.eos_token_id,
            pad_token_id=tokenizer.pad_token_id,
        )
    # í”„ë¡¬í”„íŠ¸ ì´í›„ ìƒì„± í† í°ë§Œ ì¶”ì¶œ
    gen_ids = outputs[0][inputs["input_ids"].shape[1]:]
    text_out = tokenizer.decode(gen_ids, skip_special_tokens=True)
    return text_out.strip()

# llm = RunnableLambda(hf_generate)
llm = RunnableLambda(hf_generate_direct)


# 4) ë²¡í„°ìŠ¤í† ì–´/ë¦¬íŠ¸ë¦¬ë²„ (ê¸°ì¡´ db ì‚¬ìš© ê°€ì •)
# db = Chroma(persist_directory="...")  # ì´ë¯¸ êµ¬ì„±ë¼ ìˆë‹¤ë©´ ì¬ì‚¬ìš©
# retriever = db.as_retriever(search_type="mmr", search_kwargs={"k": 4, "fetch_k": 20})
retriever = db.as_retriever(search_type="similarity", search_kwargs={"k": 3})


# 5) ë¬¸ì„œ ê²°í•© ìœ í‹¸
def format_docs(docs):
    return "\n\n".join(d.page_content for d in docs)

# 6) í”„ë¡¬í”„íŠ¸ (ChatPromptTemplate ëŒ€ì‹  ìˆœìˆ˜ PromptTemplate â†’ ë¬¸ìì—´ë¡œ í¬ë§·)
prompt = PromptTemplate.from_template(
    "ë‹¤ìŒ ì»¨í…ìŠ¤íŠ¸ë§Œ ì‚¬ìš©í•´ ê°„ê²°í•˜ê²Œ í•œêµ­ì–´ë¡œ ë‹µí•˜ì„¸ìš”.\n"
    "===\n{context}\n===\nì§ˆë¬¸: {input}"
)

# 7) RAG ì²´ì¸ ì¡°ë¦½ (Runnable ì»´í¬ì§€ì…˜)
# ğŸ”§ StringPromptValue/ChatPromptValue â†’ str ë¡œ ë³€í™˜
to_str = RunnableLambda(lambda x: x.to_string() if hasattr(x, "to_string") else str(x))

rag_chain = (
    {
        "context": retriever | RunnableLambda(format_docs),
        "input": RunnablePassthrough(),
    }
    | prompt
    | to_str      # â† ì—¬ê¸°ì„œ ìˆœìˆ˜ ë¬¸ìì—´ë¡œ ë³€í™˜
    | llm
)
```

```python
preds = []

for q in tqdm(test['Question'], desc="Inference"):
    prompt = rag_chain.invoke(q)
    print("###### Question ######")
    print(q)
    print("###### Answer ######")
    print(prompt)
    pred_answer = extract_answer_only(prompt, original_question=q)
    preds.append(pred_answer)
```

```python
sample_submission = pd.read_csv('../submission/sample_submission.csv')
sample_submission['Answer'] = preds
sample_submission.to_csv('../submission/baseline_submission.csv', index=False, encoding='utf-8-sig')
```
