```python
from PyPDF2 import PdfReader
import re
import json

# 1. PDF 로드
pdf_path = "../data/개인정보 보호법.pdf"
reader = PdfReader(pdf_path)
```

```python

# 2. 전체 텍스트 추출
full_text = ""
for page in reader.pages:
    full_text += page.extract_text() + "\n"

print(full_text)
```

```python
output_path = '정제된_개인정보보호법.txt'

# 조문 패턴(예: 제1조(목적) ...)
pattern = r'제\d+조\s*\(.*?\)[\s\S]*?(?=제\d+조|\Z)'
matches = re.findall(pattern, full_text)

# 결과 저장 및 출력
with open(output_path, 'w', encoding='utf-8') as f:
    for match in matches:
        f.write(match.strip() + '\n\n')
```

```python
# 첫 5개 결과 예시 출력
for m in matches[:5]:
    print(m.strip())
    print('------')
```

```python
cleaned_matches = []
for m in matches:
    # 줄바꿈 제거
    s = m.replace('\n', ' ')
    # <개정 yyyy. mm. dd.> 형식 문자열 제거
    s = re.sub(r'<[^>]*>', '', s)
    # 한자 제거 (유니코드 범위: \u4E00-\u9FFF)
    s = re.sub(r'[\u4E00-\u9FFF]', '', s)
    # 불필요한 다중 공백 정리
    s = re.sub(r'\s+', ' ', s).strip()
    # () 문자열 패턴 정리
    s = re.sub(r'\(\s*\)', '', s)
    cleaned_matches.append(s)

# 결과 예시 5개 출력
for c in cleaned_matches[:5]:
    print(c)
    print('------')
```

```python
import re

def extract_structure(text):
    """
    장/절/조/항 구조 추출 함수.
    """
    chapter = re.search(r'(제\d+장\s*[^\s]*)', text)
    section = re.search(r'(제\d+절\s*[^\s]*)', text)
    article = re.search(r'(제\d+조\s*\([^\)]+\))', text)
    return (
        chapter.group(1) if chapter else None,
        section.group(1) if section else None,
        article.group(1) if article else None
    )

def clean_text(s):
    # 불필요한 꼬리표 및 메타 텍스트 제거
    s = re.sub(r'법제처\s*\d*\s*국가법령정보센터\s*개인정보\s*보호법', '', s)
    s = re.sub(r'국가법령정보센터\s*개인정보\s*보호법', '', s)
    # <...> 형태 삭제
    s = re.sub(r'<[^>]*>', '', s)
    # 한자 제거
    s = re.sub(r'[\u4E00-\u9FFF]', '', s)
    # 문단 앞뒤 공백 정리
    s = re.sub(r'\s+', ' ', s).strip()
    # 빈 괄호 제거
    s = re.sub(r'\(\s*\)', '', s)
    return s

def split_paragraphs(s):
    """
    ①, ②, ③, ... 또는 번호가 붙은 항목으로 문단 분리(최대한 보존)
    """
    paras = re.split(r'(?=[①-⑳]\s*)', s)
    # 빈 문장 제거
    return [p.strip() for p in paras if p.strip()]

# 조문 중복 체크용
seen_articles = set()
processed = []

for m in matches:
    s = m.replace('\n', ' ')
    s = clean_text(s)
    # 구조 정보 추출
    chapter, section, article = extract_structure(s)
    # 조문 중복 방지
    if article and article in seen_articles:
        continue
    if article:
        seen_articles.add(article)
    # 문단 분리
    paragraphs = split_paragraphs(s)
    # 최종 구조화
    obj = {
        "chapter": chapter,
        "section": section,
        "article": article,
        "paragraphs": paragraphs
    }
    processed.append(obj)
```

```python
qa_pairs = []
for item in processed:
    article = item['article'] or ""
    for p in item['paragraphs']:
        qa_pairs.append({
            "instruction": f"{article}에 대해 설명하라.",
            "input": "",
            "output": p
        })

with open('train.jsonl', 'w', encoding='utf-8') as f:
    for pair in qa_pairs:
        f.write(json.dumps(pair, ensure_ascii=False) + '\n')
```

```python
import re

def extract_article_paragraphs(s):
    # '제n조(제목) 본문'에서 본문을 ①, ②, ③ 등 기준으로 분할
    match = re.match(r'(제\d+조\([^\)]+\))\s*(.*)', s)
    if not match:
        return None, [s]
    article = match.group(1)
    content = match.group(2)
    # 문단 분리 (①, ②, ... 기준)
    paras = re.split(r'(?=(?:[①-⑳]))', content)
    paras = [p.strip() for p in paras if p.strip()]
    # 만약 문단이 없으면 전체 본문을 0번 항으로
    if not paras:
        return article, [content.strip()]
    return article, paras

def build_qa_list(processed, law_name="개인정보 보호법"):
    qa_list = []
    for item in processed:
        article = item['article']
        paragraphs = item['paragraphs']
        # 조문제목(제3조(개인정보 보호 원칙)) 단독 문장은 항목 없음
        if len(paragraphs) == 1 and (paragraphs[0].startswith(article) or re.match(r'제\d+조\(', paragraphs[0])):
            qa_list.append({
                "instruction": f"{law_name} {article}에 대해 설명하라.",
                "input": "",
                "output": paragraphs[0]
            })
        else:
            # ①, ②, ... 문단별 생성
            for idx, para in enumerate(paragraphs):
                # 본문 앞에 ①, ② 등 항목이 붙어있는지 체크
                match = re.match(r'^([①-⑳])\s*(.*)', para)
                if match:
                    num = match.group(1)
                    text = match.group(2).strip()
                    qa_list.append({
                        "instruction": f"{law_name} {article} 제{idx+1}항에 대해 설명하라.",
                        "input": "",
                        "output": f"{num} {text}" if text else num
                    })
                else:
                    # 번호 없는 일반 문단
                    qa_list.append({
                        "instruction": f"{law_name} {article}에 대해 설명하라.",
                        "input": "",
                        "output": para
                    })
    return qa_list

# ---- 아래는 예시 입력 ----
# processed = [...위에서 구조화한 리스트...]

# 예시 QA pair 생성
qa_list = build_qa_list(processed)

# 저장 예시 (jsonl)
import json
with open('privacy_law_train.jsonl', 'w', encoding='utf-8') as f:
    for q in qa_list:
        f.write(json.dumps(q, ensure_ascii=False) + '\n')

# 결과 예시 5개 출력
for q in qa_list[:5]:
    print(q)
```

```python
def is_meaningful_pair(q):
    # output이 instruction 조문명 또는 그 일부만 있는 경우 제거
    out = q['output'].strip()
    ins = q['instruction'].strip()
    # 조문명 추출(예: "제3조(개인정보 보호 원칙)")
    m = re.search(r'(제\d+조\([^)]+\))', ins)
    if m and (out == m.group(1) or out in ins):
        return False
    # output 길이가 너무 짧은 경우(불완전한 항, 숫자만 등)
    if len(out) < 10:
        return False
    # 정상 쌍
    return True

filtered_qa_list = [q for q in qa_list if is_meaningful_pair(q)]

# 결과 예시 10개만 출력
for q in filtered_qa_list[:10]:
    print(q)
```

```python
def is_strong_meaningful_pair(q):
    out = q['output'].strip()
    ins = q['instruction'].strip()

    # 너무 짧은 output
    if len(out) < 10:
        return False

    # 조문명만 있거나 instruction에 포함된 조문명만 있는 경우
    m = re.search(r'(제\d+조\([^)]+\))', ins)
    if m and (out == m.group(1) or out in ins):
        return False

    # output이 '①', '②', '③', '④', '⑤', ...로 시작하면서, 
    # 기관명(분쟁조정위원회, 보호위원회, 국가, 지방자치단체 등) + "는/은/이/가" 만 있는 경우
    short_intro = re.match(r'^[①-⑳]\s*(분쟁조정위원회|보호위원회|국가|지방자치단체|공공기관|공공단체|위원회|정보주체|수탁자|위탁자|개인정보처리자|위원장|정부|공직선거법)[\s,는은이가]*$', out)
    if short_intro:
        return False

    # output이 '제n조(조문제목)'로 시작하면서, 바로 기관명이나 "한다" 등 실질 설명이 없을 때
    m2 = re.match(r'제\d+조\([^)]+\)\s*(분쟁조정위원회|보호위원회|국가|지방자치단체|공공기관|위원회|정보주체|수탁자|위탁자|개인정보처리자|위원장|정부)?[\s,는은이가]*$', out)
    if m2:
        return False

    # output이 항번호만 있거나 한글자짜리 불완전 output
    if re.fullmatch(r'[①-⑳]', out):
        return False

    # 너무 짧거나, '한다', '있다', '개정한다', '삭제' 등 불완전 문장도 걸러줌
    last_words = ["한다.", "있다.", "없다.", "개정한다.", "삭제", "등", "등."]
    if out in last_words or out in ["삭제"]:
        return False

    # 정상적인 설명, 항목 등
    return True

filtered_qa_list = [q for q in qa_list if is_strong_meaningful_pair(q)]

# 결과 10개 예시 출력
for q in filtered_qa_list[:10]:
    print(q)
```
